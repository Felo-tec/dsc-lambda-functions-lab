## Simple arithmetic

Use a lambda function to create a new column called `'stars_squared'` by squaring the stars column.

```python
# multiple conditions with lambda function (stars column)
df['stars_squared'] = df['stars'].apply(lambda x: x**2)
df.head()
    

```

## Dates
Select the month from the date string using a lambda function.

```python
#selecting the month from the date string using lambda function
df['month'] = df['date'].apply(lambda x:x[5:7])
df.head()

```

## What is the average number of words for a yelp review?
Do this with a single line of code.

```python
#average number of words for a yelp review
df['avg_word_number'] = df['text'].apply(lambda x: sum(len(word) for word in x.split())/len(x.split()))
df.head()
```

## Create a new column for the number of words in the review

```python
#creating a new column for the numbwer of words in a review
df['word_count'] = df['text'].apply(lambda x: len(x.split()))
df[['text', 'word_count']].head()
```
